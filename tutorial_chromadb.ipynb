```python
import chromadb

client = chromadb.Client()
collection = client.get_or_create_collection(name="my_collection")
```

```python
collection.add(
    documents=["This document is about AI.","This document is about Python."],
    ids=["id1","id2"]
)
```

```python
all_docs = collection.get()
print(all_docs)
```

```python
documents = collection.get(ids=["id1"])
print(documents)    
```

```python
results = collection.query(
    query_texts=["Tell me about AI."],
    n_results=1
)   
print(results)
```

```python
collection.delete(ids=all_docs["ids"])
collection.get()  # Should return empty results
```

```python
collection.add(
    documents=["This document is about AI.","This document is about Python."],
    ids=["id3","id4"],
    metadatas=[{"url":"http://en.wikipedia.org/wiki/Artificial_intelligence"},{"url":"http://en.wikipedia.org/wiki/Python_(programming_language)"}]
)
```

```python
results = collection.query(
    query_texts=["Tell me about AI."],  
    n_results=1,
    include=["metadatas"]
)   
print(results)
```
